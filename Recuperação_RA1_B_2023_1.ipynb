1. $$\lim_{{x \to 0}} \frac{1 - \cos(2x)}{x^2}$$

a) Solução Algébrica:
Para encontrar o limite, podemos utilizar uma série de identidades trigonométricas. Vamos começar expandindo o termo cos(2x) usando a fórmula de duplicação do cosseno:

$$cos(2x) = cos^2(x) - sen^2(x)$$

Agora, substituindo essa expressão no limite original, temos:

$$\lim_{{x \to 0}} \frac{1 - \cos(2x)}{x^2} \lim_{{x \to 0}} \frac{1 - cos^2(x) + sen^2(x)}{x^2}$$

A próxima etapa é utilizar a identidade trigonométrica sen^2(x) = 1 - cos^2(x) para simplificar a expressão ainda mais:

$$\lim_{{x \to 0}} \frac{1 - cos^2(x) + sen^2(x)}{x^2} = \lim_{{x \to 0}} \frac{2sen^2(x)}{x^2}$$

Agora, podemos simplificar a expressão dividindo o numerador e o denominador por x^2:

$$\lim_{{x \to 0}} \frac{2sen^2(x)}{x^2} = 2 * \lim_{{x \to 0}} \frac{2sen^2(x)}{x^2}$$

Agora, vamos aplicar a definição de limite para simplificar ainda mais a expressão:

$$\lim_{{x \to 0}} \frac{2sen^2(x)}{x^2} = \lim_{{x \to 0}} \frac{sen^2(x)}{x^2}$$

Sabemos que \lim_{{x \to 0}} (sen(x)/x) = 1, então temos:

$$\lim_{{x \to 0}} (sen(x)/x)^2 = 1^2 = 1$$

Portanto, o limite de $$\frac{1 - cos(2x)}{x^2}$$ quando x tende a 0 é igual a 1.

```python
# b) Solução gráfica:
# Para solucionar graficamente, podemos plotar o gráfico da função y = (1 - cos(2x))/x^2 e observar o comportamento quando x se aproxima de 0.

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 1000)
y = (1 - np.cos(2*x))/x**2

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de y = (1 - cos(2x))/x^2')
plt.axvline(0, color='black', linewidth=0.5)
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.show()
```

1. $$\lim_{{x \to \infty}} \frac{3x^2 - 2x + 1}{2x^2 + x - 1}$$

a) Solução Algébrica: Para encontrar o limite, podemos dividir todos os termos da função por x^2, já que os termos de maior grau terão maior influência no limite quando x tende ao infinito:

$$\lim_{{x \to \infty}} \frac{3x^2 - 2x + 1}{2x^2 + x - 1} = \lim_{{x \to \infty}} \frac{3 - (\frac{2}{x}) + (\frac{1}{x^2})}{2 + (\frac{1}{x}) - (\frac{1}{x^2})}$$

Agora, podemos aplicar a definição de limite para simplificar a expressão:

$$\lim_{{x \to \infty}} \frac{3 - (\frac{2}{x}) + (\frac{1}{x^2})}{2 + (\frac{1}{x}) - (\frac{1}{x^2})} = \frac{3 - 0 + 0}{2 + 0 - 0} = \frac{3}{2} = 1.5$$

Portanto, o limite de $$\frac{3x^2 - 2x + 1}{2x^2 + x - 1}$$ quando x tende ao infinito é igual a 1.5.

```python
# b) Solução gráfica:
# Para solucionar graficamente, podemos plotar o gráfico da função y = (3x^2 - 2x + 1)/(2x^2 + x - 1) e 
# observar o comportamento quando x se aproxima de valores muito grandes.

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
y = (3*x**2 - 2*x + 1)/(2*x**2 + x - 1)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de y = (3x^2 - 2x + 1)/(2x^2 + x - 1)')
plt.axvline(0, color='black', linewidth=0.5)
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.show()
```

3. $$\lim_{{x \to 0}} x \cdot \ln(1 + \sin(\frac{1}{x}))$$

a) Solução Algébrica: Para resolver esse limite, vamos utilizar uma propriedade do logaritmo natural:

$$\lim_{{x \to 0}} x \cdot \ln(1 + \sin(\frac{1}{x})) = \lim_{{x \to 0}} \frac{\ln(1 + \sin(\frac{1}{x}))}{(\frac{1}{x})}$$

Agora, vamos aplicar a definição de limite e a regra de L'Hôpital para simplificar a expressão:

$$\lim_{{x \to 0}} \frac{\ln(1 + \sin(\frac{1}{x}))}{(\frac{1}{x})} = \lim_{{x \to 0}} \frac{1}{(1 + \sin(\frac{1}{x}))} * \frac{\frac{d}{dx}1 + sin(1/x)}{\frac{d}{dx}(\frac{1}{x})}$$

A derivada de (1 + sin(1/x)) em relação a x é:

$$\frac{d}{dx}1 + \sin(\frac{1}{x}) = (\frac{1}{x^2}) * \cos(\frac{1}{x})$$

E a derivada de (1/x) em relação a x é:

$$\frac{d}{dx}(\frac{1}{x}) = \frac{-1}{x^2}$$

Substituindo essas derivadas na expressão:

$$\lim_{{x \to 0}} \frac{1}{1 + sin(\frac{1}{x})} * \frac{\frac{d}{dx}1 + sin(\frac{1}{x})}{\frac{d}{dx}(\frac{1}{x})} = \lim_{{x \to 0}} \frac{1}{1 + sin(\frac{1}{x})} * \frac{(\frac{1}{x^2}) * \cos(\frac{1}{x})}{(\frac{-1}{x^2})}$$

Agora, podemos simplificar a expressão:

$$\lim_{{x \to 0}} \frac{1}{1 + sin(\frac{1}{x})} * \frac{(\frac{1}{x^2}) * \cos(\frac{1}{x})}{(\frac{-1}{x^2})} = \lim_{{x \to 0}} \frac{-\cos(\frac{1}{x})}{1 + \sin(\frac{1}{x})}$$

Agora, vamos aplicar a definição de limite para simplificar ainda mais a expressão:

$$\lim_{{x \to 0}} \frac{-\cos(\frac{1}{x})}{1 + \sin(\frac{1}{x})} = \frac{-\cos(\frac{1}{0})}{1 + \sin(\frac{1}{0})}$$

No entanto, os valores 1/0 e sin(1/0) não são definidos, portanto, não podemos determinar o limite dessa forma.

# Questão 2
Considerando a função definida por: 
A função descontínua $f(x)$ é definida como:

$$
f(x) = 
\begin{cases} 
\sin(x) & \text{se } -10 \leq x < -3 \\
\cos(x) & \text{se } -3 \leq x < 0 \\
x^2 & \text{se } 0 \leq x < 3 \\
e^{x-10} & \text{se } 3 \leq x \leq 10
\end{cases}
$$

Trace o gráfico desta função, marque os pontos de descontinuidade, caso existam estes pontos, com um círculo vermelho, coloque o valor destes pontos. Não se esqueça de colocar uma legenda que permita o entendimento do seu gráfico indicando as curvas correspondentes a cada equação. 

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-10, -3, 1000)
y1 = np.sin(x1)

x2 = np.linspace(-3, 0, 1000)
y2 = np.cos(x2)

x3 = np.linspace(0, 3, 1000)
y3 = x3**2

x4 = np.linspace(3, 10, 1000)
y4 = np.exp(x4-10)

plt.plot(x1, y1, label='sin(x)', color='blue')
plt.plot(x2, y2, label='cos(x)', color='green')
plt.plot(x3, y3, label='x^2', color='purple')
plt.plot(x4, y4, label='e^(x-10)', color='orange')

# Plotar os pontos de descontinuidade
plt.plot(-3, np.cos(-3), 'ro', markersize=8, label='Descontinuidade em x=-3')
plt.plot(0, 0, 'ro', markersize=8, label='Descontinuidade em x=0')
plt.plot(3, np.exp(3-10), 'ro', markersize=8, label='Descontinuidade em x=3')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função descontínua')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
