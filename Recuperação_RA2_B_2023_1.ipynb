## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$

Para derivar essa função, podemos aplicar a regra da potência e a regra da soma para cada termo:

$$f(x) = \frac{d}{dx}(x^3) + \frac{d}{dx}(5x^2) - \frac{d}{dx}(7x) + \frac{d}{dx}(9)$$

Aplicando a regra da potência, temos:

$$f(x) = 3x^2 + 2 * 5x - 1 * 7 + 0$$

Simplificando, temos:

$$f(x) = 3x^2 + 10x - 7$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
f_x = x**3 + 5*x**2 - 7*x + 9
f_prime_x = 3*x**2 + 10*x - 7

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, f_x, label='f(x)', color='blue')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = x^3 + 5x^2 - 7x + 9')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, f_prime_x, label="f'(x)", color='green')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title('Gráfico da derivada de f(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

## 2.
$$g(x) = e^x \cos(x)$$

Para derivar essa função, vamos aplicar a regra do produto e a regra da cadeia:

$$g(x) = (\frac{d}{dx}(e^x)) * cos(x) + e^x * (\frac{d}{dx}(cos(x)))$$

A derivada de e^x em relação a x é simplesmente e^x:

$$\frac{d}{dx}(e^x) = e^x$$

A derivada de cos(x) em relação a x é -sin(x):

$$\frac{d}{dx}(\cos(x)) = -\sin(x)$$

Agora, podemos substituir essas derivadas na expressão da derivada de g(x):

$$g(x) = e^x * \cos(x) + e^x * (-\sin(x))$$

Simplificando, temos:

$$g(x) = e^x * (\cos(x) - \sin(x))$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 1000)
g_x = np.exp(x) * np.cos(x)
g_prime_x = np.exp(x) * (np.cos(x) - np.sin(x))

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, g_x, label='g(x)', color='blue')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Gráfico de g(x) = e^x * cos(x)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, g_prime_x, label="g'(x)", color='green')
plt.xlabel('x')
plt.ylabel("g'(x)")
plt.title('Gráfico da derivada de g(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

##2. 
$$j(x) = \sqrt{x} \ln(x)$ para $x=3$$

Para calcular a derivada de j(x), aplicaremos a regra do produto e a regra da cadeia:

$$j(x) = (\frac{d}{dx}(\sqrt{x})) * ln(x) + \sqrt{x} * (\frac{d}{dx}(ln(x)))$$

A derivada de sqrt(x) em relação a x pode ser encontrada usando a regra do quociente:

$$\frac{d}{dx}(\sqrt{x}) = \frac{1}{2} * (1/\sqrt{x}) = \frac{1}{2\sqrt{x}}$$

A derivada de ln(x) em relação a x é 1/x:

$$\frac{d}{dx}(ln(x)) = \frac{1}{x}$$

Substituindo essas derivadas na expressão da derivada de j(x):

$$j(x) = \frac{1}{2\sqrt{x}} * ln(x) + \sqrt{x} * \frac{1}{x}$$

Simplificando, temos:

$$j(x) = \frac{ln(x) + 2\sqrt{x}}{2\sqrt{x}}$$

Agora, podemos substituir x=3 para encontrar o valor da derivada nesse ponto:

$$j(3) = \frac{ln(3) + 2\sqrt{3}}{2\sqrt{3}}$$

```python
j_prime_3 = (math.log(3) + 2*math.sqrt(3))/(2*math.sqrt(3))
j_prime_3
```

# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
import numpy as np

def f(x):
    return x**2 - 98*x + 4

x = np.linspace(0, 100, 1000)
y = f(x)

max_index = np.argmax(y)
max_x = x[max_index]
max_y = y[max_index]

print("O máximo local da função ocorre em x =", max_x, "e y =", max_y)
```
