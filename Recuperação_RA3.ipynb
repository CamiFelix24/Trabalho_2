#Questão 1 (vale até 2 pontos)
Determine a área entre as curvas definidas por $x_1=y^2-2$ e $x_2=y$. Trace o gráfico destas funções sombreando em azul claro a área entre elas. Você pode usar apenas o python, matplotlib, numpy e sympy. Neste caderno do Google Colab.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x, y = sp.symbols('x y')

curve1 = y - 2
curve2 = y

x1 = sp.solve(curve1 - x, x)[0]
x2 = sp.solve(curve2 - x, x)[0]

x1_func = sp.lambdify(y, x1, modules=['numpy'])
x2_func = sp.lambdify(y, x2, modules=['numpy'])

y_values = np.linspace(0, 5, 100)
x1_values = x1_func(y_values)
x2_values = x2_func(y_values)

area = sp.integrate(curve2 - curve1, (y, 0, 5))

fig, ax = plt.subplots()
ax.plot(x1_values, y_values, label='x = y - 2')
ax.plot(x2_values, y_values, label='x = y')
ax.fill_betweenx(y_values, x1_values, x2_values, where=(x1_values <= x2_values), color='lightblue')
ax.legend()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Área entre as curvas')
plt.grid(True)
plt.show()

print("A área entre as curvas é aproximadamente:", sp.N(area))
```

Usando apenas álgebra, encontre o valor exato da área entre as seguintes funções: $y_1=e^x$, $y_2=x^{2} -1$, $x_1=-1$ e $x_2=1$. Lembre-se o cálculo precisa ser detalhado, e impresso em qualidade matemática (usando latex) neste caderno do Google Colab.

##Resposta


##$y_1=e^x$, $y_2=x^{2} -1$, $x_1=-1$ e $x_2=1$

Podemos calcular a integral definida da diferença $y_2 - y_1$ no intervalo dado.

$$Área = \int_{-1}^1 (y_2 - y_1) \space dx$$
$$ = \int_{-1}^1 (x^2 - 1 - e^x) \space dx$$

Podemos encontrar essa integral calculando a primitiva da função $x^2 - 1 - e^x$ em relação a $x$ e, em seguida, avaliando-a nos limites de integração. Vamos calcular a primitiva:

$$ = \int (x^2 - 1 - e^x) \space dx = \frac{1}{3}x^3 - x - e^x + C$$

Agora, vamos calcular o valor da área substituindo os limites de integração:

$$Área = [\frac{1}{3}x^3 - x - e^x]_{-1}^1$$
$$= (\frac{1}{3}(1)^3 - 1 - e^1) - (\frac{1}{3}(-1)^3 - (-1) - e^(-1))$$
$$= (\frac{1}{3} - 1 - e) - (-\frac{1}{3} + 1 - \frac{1}{e})$$
$$= \frac{1}{3} - 1 - e + \frac{1}{3} - 1 + \frac{1}{e}$$
$$= \frac{2}{3} - 2 - e + \frac{1}{e}$$

Portanto, o valor exato da área entre as funções $y_1 = e^x$ e $y_2 = x^2 - 1$ no intervalo $[-1,1]$ é $\frac{2}{3} - 2 - e + \frac{1}{e}$.

#Questão 3 (vale até 2 pontos)
Um objeto está se movendo e experimenta diferentes acelerações em diferentes tempos. Os tempos e as respectivas acelerações são:

1. No instante $t = 1$ segundo, aceleração $a = 4 m/s²$
2. No instante $t = 3$ segundos, aceleração $a = -2 m/s²$
3. No instante $t = 4$ segundos, aceleração $a = 6 m/s²$
4. No instante $t = 6$ segundos, aceleração $a = 1 m/s²$
A posição inicial do objeto é $s_0 = 0 m$ e a velocidade inicial é $v_0 = 0 m/s$.

Você precisa imprimir com qualidade matemática (latex) a função encontrada para a aceleração, a função encontrada para a posição e a posição no instante $t=4.2s$. Além, é claro, dos gráficos de aceleração e posição.

Você poderá usar apenas as bibliotecas matplotlib, numpy e sympy.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

t_values = [1, 3, 4, 6]
a_values = [4, -2, 6, 1]

t = sp.symbols('t')

a_func = sp.Piecewise((a_values[0], t < t_values[0]),
                      (a_values[1], t < t_values[1]),
                      (a_values[2], t < t_values[2]),
                      (a_values[3], True))

v_func = sp.integrate(a_func, t)
s_func = sp.integrate(v_func, t)

a_numeric = sp.lambdify(t, a_func, modules=['numpy'])
s_numeric = sp.lambdify(t, s_func, modules=['numpy'])

t_calc = 4.2

s_calc = s_numeric(t_calc)

t_plot = np.linspace(0, t_values[-1], 100)
a_plot = a_numeric(t_plot)

fig, ax1 = plt.subplots()
ax1.plot(t_plot, a_plot, color='red')
ax1.set_xlabel('Tempo (s)')
ax1.set_ylabel('Aceleração (m/s²)')
ax1.set_title('Gráfico de Aceleração')
ax1.grid(True)

s_plot = s_numeric(t_plot)

ax2 = ax1.twinx()
ax2.plot(t_plot, s_plot, color='blue')
ax2.set_ylabel('Posição (m)')

ax2.plot(t_calc, s_calc, marker='o', color='green')
ax2.annotate(f't={t_calc}s, s={s_calc:.2f}m', (t_calc, s_calc),
              textcoords="offset points", xytext=(10,10), ha='center')

plt.tight_layout()
plt.show()

print('Função de Aceleração:')
print(sp.latex(a_func))

print('\nFunção de Posição:')
print(sp.latex(s_func))

print('\nPosição no instante t=4.2s:')
print(f's = {s_calc:.2f}m')
```

#Questão 5 (vale até 2 pontos)
Considere as seguintes curvas no plano cartesiano:

Curva A: $y = x^2$
Curva B: $y = 2x - 1$

Encontre os valores de [a, b], de modo que a área entre estas curvas neste intervalo seja igual a 6 unidades quadradas sabendo que $a=2$.

Usando o Python e apenas as bibliotecas matplotlib, numpy e sympy,
1. crie um código que plote estas duas curvas no intervalo entre $-10 \leq x \leq 10$;
2. crie um código que econtre o valor desejado de $b$;
3. crie um código que plote estas duas curvas sombreando em azul claro a área entre elas limitada pelo intervalo $[a,b]$ de forma que seja possível ver esta área completamente;
4. prove que o valor de b que você encontrou produzirá uma área de 6 unidades quadradas com precisão maior ou igual a 0.001.


## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x, y = sp.symbols('x y')

curve_A = x**2
curve_B = 2*x - 1

x_vals = np.linspace(-10, 10, 100)
curve_A_vals = sp.lambdify(x, curve_A)(x_vals)
curve_B_vals = sp.lambdify(x, curve_B)(x_vals)

plt.plot(x_vals, curve_A_vals, label='y = x^2')
plt.plot(x_vals, curve_B_vals, label='y = 2x - 1')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curvas y = x^2 e y = 2x - 1')
plt.grid(True)
plt.show()

a = 2

intersection_eq = sp.Eq(curve_A, curve_B)
intersections = sp.solve(intersection_eq, x)

valid_intersections = [intersection for intersection in intersections if intersection.is_real and a <= intersection <= 10]

if len(valid_intersections) > 0:
    b = valid_intersections[0]
    print('Valor de b:', b)

    x_range = np.linspace(a, b, 100)
    curve_A_range = sp.lambdify(x, curve_A)(x_range)
    curve_B_range = sp.lambdify(x, curve_B)(x_range)

    plt.plot(x_vals, curve_A_vals, label='y = x^2')
    plt.plot(x_vals, curve_B_vals, label='y = 2x - 1')
    plt.fill_between(x_range, curve_A_range, curve_B_range, where=(curve_A_range <= curve_B_range), color='lightblue')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Área entre as curvas')
    plt.grid(True)
    plt.show()

    area = sp.integrate(curve_A - curve_B, (x, a, b))

    if abs(area - 6) <= 0.001:
        print('O valor de b produz uma área de 6 unidades quadradas com precisão maior ou igual a 0.001.')
    else:
        print('O valor de b não produz uma área de 6 unidades quadradas com precisão maior ou igual a 0.001.')
else:
    print('Não foi encontrado um valor válido para b.')

```

a) Calcule a integral indefinida:

b) Calcule a integral indefinida:

c) Calcule a integral definida:

d) Calcule a integral indefinida:

## Respostas


##a) $\int (3x^2 + 2e^x) \, dx$

Integrando o termo $3x^2$, usamos a regra de potência da integral:

$$\int 3x^2 \space dx = \frac{3}{3} * \frac{x^3}{3} + C_1 = x^3 + C_1$$

Integrando o termo $2e^x$, usamos a regra de integração da função exponencial:

$$\int 2e^x \space dx = 2 * e^x + C_2$$

Portanto, a integral indefinida completa é dada por:

$$\int (3x^2 + 2e^x) \, dx = x^3 + 2e^x + C$$

onde $C = C_1 + C_2$é a constante de integração.
Assim, a solução para a integral indefinida $\int (3x^2 + 2e^x) \, dx = x^3 + 2e^x + C$

##b) $\int x \sin(x) \, dx$

Vamos aplicar a fórmula de integração por partes:

$$\int u \space dv = uv - \int v \space du$$

Escolhendo $u = x$ e $dv = \sin(x) \ space dx$, podemos determinar $du$ e $v$ da seguinte forma:

$du = dx$(derivada de $u$)

$v = -\cos(x)$ (integral de $dv$)

Aplicando a fórmula de integração por partes, obtemos:

$$\int x \sin(x) \space dx = -x \cos(x) - \int (-\cos(x)) \space dx$$

Simplificando, temos:

$$\int x \sin(x) \space dx = -x \cos(x) + \int\cos(x) \space dx$$

A integral de $\cos(x)$ é $\sin(x)$, portanto:

$$\int x \sin(x) \space dx = -x \cos(x) + \sin(x) + C$$

onde $C$ é a constante de integração.

Assim, a solução para a integral indefinida $\int x \sin(x) \space dx = -x \cos(x) + \sin(x) + C$

##c) $\int_0^{\pi} x \cos(x) \, dx$

Vamos aplicar a fórmula de integração por partes para integrais definidas:

$$\int_a^{b} u \space dv = [uv]_a^{b} - \int_a^{b} v \space du$$

Escolhendo $u = x$ e $dv = \cos(x) \space dx$, podemos determinar $du$ e $v$ da seguinte forma:

$du=dx$ (derivada de $u$)

$v = \sin(x)$(integral de $dv$)

Aplicando a fórmula de integração por partes, temos:

$$\int_0^{\pi} x \cos(x) \space dx = [x \sin(x)]_0^{\pi} - \int_0^{\pi} \sin(x) \space dx$$

Avaliando a integral no intervalo $[0,π]$, obtemos:

$$\int_0^{\pi} x \cos(x) \space dx = [x \sin(x)]_0^{\pi} - [-\cos(x)]_0^{\pi}$$

Substituindo os limites de integração, temos:

$$\int_0^{\pi} x \cos(x) \space dx = ({\pi}\sin({\pi}) - 0 \sin(0)) - (-\cos({\pi}) + \cos(0))$$

Simplificando, obtemos:

$$\int_0^{\pi} x \cos(x) \space dx = {\pi}\sin({\pi}) + \cos(\pi) - \cos(0)$$

Lembrando que $\sin(π) = 0$ e $\cos(π) = -1$, temos:

$$\int_0^{\pi} x \cos(x) \space dx = \pi * 0 + (-1) - \cos(0)$$

Simplificando ainda mais, temos:

$$\int_0^{\pi} x \cos(x) \space dx = -1 - \cos(0)$$

Lembrando que $\cos(0) = 1$, temos:

$$\int_0^{\pi} x \cos(x) \space dx = -1 -1 = -2$$

Portanto, a integral definida $\int_0^{\pi} x \cos(x)$ é igual a $-2$.

##d) $\int x^2 e^x \, dx$

Aplicando a fórmula de integração por partes:

$$\int u \space dv = uv - \int v \space du$$

Podemos escolher $u = x^2$ e $dv = e^x dx$. Calculando as derivadas e integrais correspondentes:

$du = 2x \space dx$ (derivada de $u$)

$v = e^x$ (integral de $dv$)

Aplicando a fórmula de integração por partes, temos:

$$\int x^2 e^x \space dx = x^2 e^x - \int 2xe^x \space dx$$

Agora, a integral restante $\int 2xe^x \space dx$ também pode ser resolvida utilizando integração por partes. Novamente, aplicamos a fórmula de integração por partes, mas dessa vez com $u = 2x$ e $dv = e^x \space dx$:

$du = 2x \space dx$ (derivada de $u$)

$v = e^x$ (integral de $dv$)

Aplicando a fórmula, temos:

$$\int 2xe^x \space dx = 2xe^x - \int 2e^x \space dx = 2xe^x - 2e^x$$

Substituindo esse resultado de volta na integral original:

$$\int x^2 e^x \space dx = x^2 e^x - (2xe^x - 2e^x) = x^2 e^x - 2xe^x + 2e^x + C$$

Portanto, a solução para a integral indefinida $\int x^2 e^x \space dx$ é $x^2 e^x - 2xe^x + 2e^x + C$ onde $C$ é a constante de integração.
