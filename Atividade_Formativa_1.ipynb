```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')

f = 40 * x**3 + 12 * x**2 - 9 * x

x_vals = np.linspace(2, 8, 100) 
f_vals = [sp.N(f.subs(x, val)) for val in x_vals]

plt.plot(x_vals, f_vals, label='f(x) = 40x^3 + 12x^2 - 9x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()

area_algebraica = sp.integrate(f, (x, 2, 7)) * 2.5
print('Área (álgebra):', area_algebraica)

x_vals = np.linspace(2, 7, 1000)
f_vals = [sp.N(f.subs(x, val)) for val in x_vals]
dx = (7 - 2) / 1000
area_riemann = np.sum(f_vals) * dx * 2.5
print('Área (Soma de Riemann):', area_riemann)
```

```python
t = sp.symbols('t')

g = 12 * t**7 - t**2 - t + 3

t_vals = np.linspace(8, 11, 100)
g_vals = [sp.N(g.subs(t, val)) for val in t_vals]

plt.plot(t_vals, g_vals, label='g(t) = 12t^7 - t^2 - t + 3')
plt.xlabel('t')
plt.ylabel('g(t)')
plt.legend()
plt.grid(True)
plt.show()

area_algebraica = sp.integrate(g, (t, 8, 10)) * 7.5
print('Área (álgebra):', area_algebraica)

t_vals = np.linspace(8, 10, 1000)
g_vals = [sp.N(g.subs(t, val)) for val in t_vals]
dt = (10 - 8) / 1000
area_riemann = np.sum(g_vals) * dt * 7.5
print('Área (Soma de Riemann):', area_riemann)
```

```python
x = sp.symbols('x')

h = (2 * x**5 - x + 3) / x**2

x_vals = np.linspace(0.1, 2.5, 100)
h_vals = [sp.N(h.subs(x, val)) for val in x_vals]

plt.plot(x_vals, h_vals, label='h(x) = (2x^5 - x + 3)/x^2')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.legend()
plt.grid(True)
plt.show()

area_algebraica = sp.integrate(h, (x, 0.1, 2))
print('Área (álgebra):', area_algebraica)

x_vals = np.linspace(0.1, 2, 1000)
h_vals = [sp.N(h.subs(x, val)) for val in x_vals]
dx = (2 - 0.1) / 1000
area_riemann = np.sum(h_vals) * dx
print('Área (Soma de Riemann):', area_riemann)
```

```python
x = sp.symbols('x')

k = 7 * sp.sin(x) - 2 * sp.cos(x)

x_vals = np.linspace(-1, np.pi/2 + 1, 100)
k_vals = [sp.N(k.subs(x, val)) for val in x_vals]

plt.plot(x_vals, k_vals, label='k(x) = 7sin(x) - 2cos(x)')
plt.xlabel('x')
plt.ylabel('k(x)')
plt.legend()
plt.grid(True)
plt.show()

area_algebraica = sp.integrate(k, (x, 0, np.pi/2))
print('Área (álgebra):', area_algebraica)

x_vals = np.linspace(0, np.pi/2, 1000)
k_vals = [sp.N(k.subs(x, val)) for val in x_vals]
dx = (np.pi/2 - 0) / 1000
area_riemann = np.sum(k_vals) * dx
print('Área (Soma de Riemann):', area_riemann)

```

```python
x = sp.symbols('x')

m = 5 * sp.sin(2 * x) - 3**x

x_vals = np.linspace(-np.pi - 1, np.pi + 1, 100)
m_vals = [sp.N(m.subs(x, val)) for val in x_vals]

plt.plot(x_vals, m_vals, label='m(x) = 5sin(2x) - 3^x')
plt.xlabel('x')
plt.ylabel('m(x)')
plt.legend()
plt.grid(True)
plt.show()

area_algebraica = sp.integrate(m, (x, -np.pi, np.pi))
print('Área (álgebra):', area_algebraica)

x_vals = np.linspace(-np.pi, np.pi, 1000)
m_vals = [sp.N(m.subs(x, val)) for val in x_vals]
dx = (np.pi - (-np.pi)) / 1000
area_riemann = np.sum(m_vals) * dx
print('Área (Soma de Riemann):', area_riemann)
```
