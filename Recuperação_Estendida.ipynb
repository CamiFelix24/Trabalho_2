# Questão 1 (vale 1 ponto)
Suponha que você esteja projetando uma ponte suspensa e precise determinar a deformação máxima que ocorrerá em um determinado ponto da estrutura. A partir de análises e medições preliminares, você sabe que a deformação em um ponto na posição x da ponte pode ser aproximada por uma função $f(x) = 3x^2 - 2x + 5$. Precisamos saber a deformação quando $x$ se aproxima do valor $2$.

Utilizando a Fórmula das diferenças finitas encontre os limites para cinco pontos à esquerda do ponto desejado e cinco pontos à direita do ponto desejado. Depois traçe um gráfico desta função anote o limite no ponto desejado e imprima os limites nos cinco pontos à esquerda e à direita.




## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt

def deformacao(x):
    return 3*x**2 - 2*x + 5

def limite_diferencas_finitas(f, x, h):
    limite_esquerda = (f(x) - f(x-h))/h
    limite_direita = (f(x+h) - f(x))/h
    return limite_esquerda, limite_direita

x = 2
h = 0.1

limites_esquerda = []
limites_direita = []

for i in range(5, 0, -1):
    limite_esq, limite_dir = limite_diferencas_finitas(deformacao, x - i*h, h)
    limites_esquerda.append(limite_esq)
    limites_direita.append(limite_dir)

limite_ponto_desejado = deformacao(x)

for i in range(1, 6):
    limite_esq, limite_dir = limite_diferencas_finitas(deformacao, x + i*h, h)
    limites_esquerda.append(limite_esq)
    limites_direita.append(limite_dir)

x_vals = np.linspace(x - 1.5, x + 1.5, 100)

y_vals = deformacao(x_vals)

plt.plot(x_vals, y_vals, label='Função de Deformação')

plt.scatter(x, limite_ponto_desejado, color='red', label='Limite no ponto desejado')

for i in range(5):
    plt.scatter(x - (i+1)*h, limites_esquerda[i], color='blue', label=f'Limite à esquerda {i+1}')

for i in range(5):
    plt.scatter(x + (i+1)*h, limites_direita[i], color='green', label=f'Limite à direita {i+1}')

print("Limites à esquerda:")
for limite in limites_esquerda:
    print(limite)

print("Limite no ponto desejado:")
print(limite_ponto_desejado)

print("Limites à direita:")
for limite in limites_direita:
    print(limite)

plt.xlabel('x')
plt.ylabel('Deformação')
plt.legend()
plt.title('Gráfico da Função de Deformação com Limites')
plt.grid(True)

plt.show()
```

#Questão 4 (vale 2 pontos)
Considere a função $f(x)= x^2 -1$. (a) Trace o gráfico desta função e, (b) usando a definição de derivada, encontre a inclinação da reta tangente nos pontos $(-1,0)$, $(0,-1)$ e $(2,3)$.

## Resposta Algébrica


$$f(x)= x^2 -1$$

Derivada de $f(x)$:

$f'(x) = \frac{d}{dx}(x^2 - 1)$

$f'(x) = 2x$

Substituindo $x_0$ nos pontos dados:

Ponto $(-1, 0)$:

$f′(−1)=2(−1)=−2$

A inclinação da reta tangente em $(-1, 0)$ é $2$.

Ponto $(0, -1)$:

$f′(0)=2(0)=0$

A inclinação da reta tangente em $(0, -1)$ é $0$.

Ponto $(2, 3)$:

$f′(2)=2(2)=4$

A inclinação da reta tangente em $(2, 3)$ é $4$.

## Resposta com código

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff

x = symbols('x')
f = x**2 - 1

x_vals = np.linspace(-3, 3, 100)

f_vals = [f.subs(x, val) for val in x_vals]

plt.plot(x_vals, f_vals, label='f(x) = x^2 - 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico de f(x) = x^2 - 1')
plt.legend()
plt.grid(True)
plt.show()

f_prime = diff(f, x)

points = [(-1, 0), (0, -1), (2, 3)]
for point in points:
    x_val, y_val = point
    slope = f_prime.subs(x, x_val)
    print(f'Inclinação da reta tangente no ponto ({x_val}, {y_val}): {slope}')
```

# Questão 5 (vale 1 ponto)
Você é um engenheiro de produção em uma fábrica de vidro que produz garrafas. A velocidade de produção das garrafas $V$ varia ao longo do tempo devido ao desgaste das máquinas e a fadiga dos trabalhadores. Depois de um mês de análise, você descobriu que a velocidade de produção é dada pela função $V(t) = 50 - 4t^2$ garrafas por hora, onde $t$ é o tempo em horas desde o início do turno de trabalho.
Seu gerente quer saber se há um momento específico durante o turno de trabalho em que a produção de garrafas é mais eficiente.
a) Como você poderia ajudar seu gerente a encontrar essa informação?
b) Se este momento existe, que momento é este?
c) Trace o gráfico da sua função usando apenas python, matplotlib, numpy e sympy e, caso este momento exista, marque este ponto no seu gráfico.

## Resposta  


a) Para ajudar meu gerente primeiro iria calcule a derivada da função de velocidade de produção para obter a taxa de variação, depois encontrar os pontos críticos onde a derivada se iguala a zero ou é indefinida, e por fim, avaliarei numericamente a função nos pontos críticos para determinar o momento específico de eficiência, se existir.

b) Se há um momento de eficiência durante o turno de trabalho em que a produção de garrafas é mais eficiente, ele ocorre quando a taxa de variação da produção é zero. Encontrarei os pontos críticos da função de velocidade de produção, onde a derivada se iguala a zero ou é indefinida, permite identificar esse momento específico de eficiência.

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff

t = symbols('t')
V = 50 - 4 * t**2

t_vals = np.linspace(-5, 5, 100)

V_vals = [V.subs(t, val) for val in t_vals]

plt.plot(t_vals, V_vals, label='V(t) = 50 - 4t^2')
plt.xlabel('Tempo (horas)')
plt.ylabel('Velocidade de produção (garrafas por hora)')
plt.title('Gráfico da Velocidade de Produção')
plt.legend()
plt.grid(True)

V_prime = diff(V, t)

critical_points = [t_val for t_val in np.linspace(-5, 5, 100) if abs(V_prime.subs(t, t_val)) < 1e-5]

if len(critical_points) > 0:
    efficiency_points = [(t_val, V.subs(t, t_val)) for t_val in critical_points]

    for point in efficiency_points:
        plt.plot(point[0], point[1], 'ro', label='Ponto de Eficiência')
        plt.text(point[0], point[1], f'({point[0]}, {point[1]})')

    plt.legend()

plt.show()

if len(critical_points) > 0:
    print('Momentos de eficiência:')
    for point in efficiency_points:
        print(f'- Tempo: {point[0]} horas, Velocidade: {point[1]} garrafas por hora')
else:
    print('Não foi encontrado um momento específico de eficiência durante o turno de trabalho.')
```

# Questão 6 (vale 1 ponto)
Considere a função $f(x) = x^3 - 10x^2 + sin(x)$. Usando apenas o Python, numpy, sympy e matplotlib, trace o gráfico desta função e a reta tangente ao ponto $x=4$.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sin, diff, lambdify

x = symbols('x')
f = x**3 - 10*x**2 + sin(x)

f_numeric = lambdify(x, f, modules=['numpy'])

x_vals = np.linspace(-5, 10, 500)

f_vals = f_numeric(x_vals)

plt.plot(x_vals, f_vals, label='f(x) = x^3 - 10x^2 + sin(x)')

x_point = 4

f_prime = diff(f, x)

slope = f_prime.subs(x, x_point)

y_point = f_numeric(x_point)

tangent_line = slope * (x_vals - x_point) + y_point

plt.plot(x_vals, tangent_line, label='Reta Tangente')

plt.plot(x_point, y_point, 'ro', label=f'Ponto de Tangência ({x_point}, {y_point})')
plt.text(x_point, y_point, f'({x_point}, {y_point})')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Função e Reta Tangente')
plt.legend()
plt.grid(True)

plt.show()
```

#Questão 7 (vale 1 ponto)
Considere um fenômeno físico em que a posição de um objeto em movimento varia com o tempo. A taxa de variação da posição em relação ao tempo possui um componente trigonométrico e é dada pela função $f(t) = A * cos(t)$, onde $A$ é uma constante que determina a amplitude da variação. Sua tarefa é plotar o gráfico da função da taxa de variação e da função posição deste objeto considerando que $A$ é um número inteiro que varia entre 2 e 5.

## Resposta

```python
import numpy as np
import matplotlib.pyplot as plt

t_vals = np.linspace(0, 2*np.pi, 100)

A_vals = np.arange(2, 6)

for A in A_vals:
    f_vals = A * np.cos(t_vals)

    s_vals = np.cumsum(f_vals)

    plt.plot(t_vals, f_vals, label=f'A = {A}')

    plt.plot(t_vals, s_vals, label=f'Posição (A = {A})')

plt.xlabel('Tempo')
plt.ylabel('Valor')
plt.title('Gráfico da Função da Taxa de Variação e da Função Posição')
plt.legend()
plt.grid(True)
plt.show()
```

#Questão 8 (vale 2 pontos)
Calcule a área entre as curvas $f(x) = -x^2$ e $g(x) = -x$, apenas no quarto quadrante do plano cartesiano, usando apenas álgebra e garantindo que todos os passos algébricos necessários estejam explicitados em notação matemática em uma célula de textos deste carderno de tarefas usando latex. Usando o python, matplotlib, numpy e sympy crie um gráfico destas funções sombreando a área entre elas em azul claro.

## Resposta



Achando os pontos de interseção:

$$-f(x) = -x^2$$
$$-x = -x^2$$
$$x^2 - x = 0$$
$$x(x - 1) = 0$$

Pontos de interseção: $x = 0$ e $x = 1$.

$$Área = ∫_0^{1} (g(x) - f(x))dx$$

Substituindo as funções $f(x)$ e $g(x)$:

$$Área = ∫_0^{1} (-x + x^2)dx$$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -x**2

def g(x):
    return -x

x_vals = np.linspace(0, 1, 100)

f_vals = f(x_vals)
g_vals = g(x_vals)

plt.plot(x_vals, f_vals, label='f(x) = -x^2')
plt.plot(x_vals, g_vals, label='g(x) = -x')

plt.fill_between(x_vals, f_vals, g_vals, where=(g_vals >= f_vals), color='lightblue')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Área entre f(x) e g(x)')
plt.legend()
plt.grid(True)

x_interval = x_vals[1] - x_vals[0]
area = np.sum((g_vals + f_vals) / 2 * x_interval)

print('Área entre f(x) e g(x) no quarto quadrante:', area)

plt.show()
```
